```python
import pandas as pd
from pandas_datareader import data
```

```python
start = '2019-04-30'
end = '2021-04-30'
source = 'yahoo'
```

```python
google = data.DataReader("GOOGL", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
tesla = data.DataReader("TSLA", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
microsoft = data.DataReader("MSFT", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
nvidia = data.DataReader("NVDA", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
ibm = data.DataReader("IBM", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
cisco = data.DataReader("CSCO", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
apple = data.DataReader("AAPL", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
amd = data.DataReader("AMD", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
dell = data.DataReader("DELL", start=start ,end=end, data_source=source).reset_index()[["Date", "Close"]]
```

```python
google["Stock"] = "Google"
tesla["Stock"] = "Tesla"
apple["Stock"] = "Apple"
nvidia["Stock"] = "Nvidia"
dell["Stock"] = "Dell"
ibm["Stock"] = "IBM"
cisco["Stock"] = "Cisco"
microsoft["Stock"] = "Microsoft"
amd["Stock"] = "Amd"
stocks = pd.concat([google, tesla, apple, nvidia, dell, ibm, cisco, microsoft, amd])
stocks["Month"] = stocks.Date.dt.month
stocks["Year"] = stocks.Date.dt.year
```
